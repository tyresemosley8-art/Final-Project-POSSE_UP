What I'm Trying to Figure Out

My group is trying to figure out what the best neighborhood in Pittsburgh is. For my part, I decided to look at two things: how affordable the housing is and how easy it is to get around without a car.

I picked these two because I think when looking for a neighborhood these are the first things I'd think about if I was actually trying to pick a neighborhood to live in. Can I afford it? And can I get around easily? If the answer to both is yes, that neighborhood is probably pretty good.

My plan is to score each neighborhood on both of those things, then combine them into one score so I can rank all the neighborhoods and pick the best one.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Now I'll load my data. My datasets are stored in an Excel file with separate sheets, so I just need to tell pandas which sheet to pull from each time.

```python
df_cost = pd.read_excel('DataSets.xlsx', sheet_name='Cost of living')
df_trans = pd.read_excel('DataSets.xlsx', sheet_name='Transportation')

print('Cost of Living rows:', len(df_cost))
print('Transportation rows:', len(df_trans))
```

---
## Part 1: Cost of Living

### First Look at the Data

Before I do anything I want to just look at the data and understand what I'm working with. I'll use `.head()` to preview the first few rows.

```python
df_cost.head(10)
```

### My Thinking

Okay so looking at this, the dataset breaks housing units into four rent brackets — under $500, $500-999, $1,000-1,999, and $2,000+. 

My idea was: if I want to know how affordable a neighborhood is, I should look at what percentage of its housing falls in the lower brackets. I decided to count anything **under $1,000/month** as affordable, since that covers the two cheapest categories.

So my affordability score is just: how many units are under $1,000 divided by the total units. Simple ratio. A neighborhood where most housing is cheap gets a score close to 1, and a neighborhood where most housing is expensive gets a score close to 0.

I also noticed one neighborhood had 0 total units which would cause a error, so I removed them first.

```python
# Remove any neighborhoods with no housing data
df_cost = df_cost[df_cost['Total Units'] > 0]

# Combine the two cheapest rent brackets
df_cost['Affordable Units'] = df_cost['< USD 500'] + df_cost['USD 500-999']

# Calculate the affordability score
df_cost['Affordability Score'] = df_cost['Affordable Units'] / df_cost['Total Units']

# Sort so the most affordable neighborhoods are at the top
df_cost = df_cost.sort_values('Affordability Score', ascending=False).reset_index(drop=True)

df_cost[['Neighborhood', 'Total Units', 'Affordable Units', 'Affordability Score']].head(10)
```

### What I Found

Some neighborhoods have nearly all of their housing under $1,000/month, which makes them really affordable. I was curious what the other end looked like too so I looked at the less affordable neighborhoods as well.

```python
# Bottom 10 - least affordable
df_cost[['Neighborhood', 'Total Units', 'Affordable Units', 'Affordability Score']].tail(10)
```

That's a pretty big difference between the top and bottom so im going to make a chart so it's easier to see.

```python
top15 = df_cost.head(15)

plt.figure(figsize=(10, 6))
plt.barh(top15['Neighborhood'], top15['Affordability Score'], color='steelblue')
plt.xlabel('Affordability Score (% of units under $1,000/month)')
plt.title('Top 15 Most Affordable Pittsburgh Neighborhoods')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

---
## Part 2: Transportation

### First Look at the Data

Now let me look at the transportation dataset.

```python
df_trans.head(10)
```

### My Thinking

This dataset shows what proportion of people in each neighborhood use different ways to get to work. There's driving alone, carpooling, public transit, biking, walking, and working from home.

My thought was a neighborhood with good transportation access is one where you have real options besides just driving. So I added up the proportions for public transit, biking, walking, and working from home. The higher that combined number, the more options people have, which to me means better transportation access.

I also capped the score at 1.0 just to make things easy in case any values were slightly over due to rounding.

```python
# Add up all the non-car commute options
df_trans['Transit Score'] = df_trans['Public Trans'] + df_trans['Bicycle'] + df_trans['Walk'] + df_trans['WFH']

# Cap at 1.0 just in case
df_trans['Transit Score'] = df_trans['Transit Score'].clip(0, 1)

# Sort highest to lowest
df_trans = df_trans.sort_values('Transit Score', ascending=False).reset_index(drop=True)

df_trans[['Neighborhood', 'Public Trans', 'Bicycle', 'Walk', 'WFH', 'Transit Score']].head(10)
```

### What I Found

Some neighborhoods have really high scores because a lot of people walk or take the bus. Also I made a stacked bar so I can see which transportation options are more popular in which neighborhoods.

```python
top15_t = df_trans.head(15)

fig, ax = plt.subplots(figsize=(10, 6))

modes = ['Public Trans', 'Walk', 'Bicycle', 'WFH']
colors = ['#2196F3', '#4CAF50', '#FF9800', '#9C27B0']
bottom = [0] * len(top15_t)

for mode, color in zip(modes, colors):
    ax.barh(top15_t['Neighborhood'], top15_t[mode], left=bottom, color=color, label=mode)
    bottom = [b + v for b, v in zip(bottom, top15_t[mode])]

ax.set_xlabel('Proportion of Commuters (non-car modes)')
ax.set_title('Top 15 Pittsburgh Neighborhoods by Transportation Access')
ax.invert_yaxis()
ax.legend(loc='lower right')
plt.tight_layout()
plt.show()
```

---
## Part 3: Putting It All Together

### Combining the Two Scores

Now I want to combine both scores into one final score for each neighborhood. I decided to weight them so equally it would be 50% affordability, 50% transportation because I think both matter just as much as the other. I could have weighted one higher but I didn't really have a strong reason to, so equal felt fair.

To do this I merge the two datasets together by neighborhood name and then average the two scores.

```python
# Grab just what I need from each
cost_slim = df_cost[['Neighborhood', 'Affordability Score']]
trans_slim = df_trans[['Neighborhood', 'Transit Score']]

# Merge by neighborhood name
df_final = pd.merge(cost_slim, trans_slim, on='Neighborhood')

# Average the two scores
df_final['My Score'] = (df_final['Affordability Score'] + df_final['Transit Score']) / 2

# Sort by final score
df_final = df_final.sort_values('My Score', ascending=False).reset_index(drop=True)

df_final.head(15)
```

```python
top15_final = df_final.head(15)

plt.figure(figsize=(10, 6))
plt.barh(top15_final['Neighborhood'], top15_final['My Score'], color='mediumseagreen')
plt.xlabel('Combined Score (Affordability + Transit averaged)')
plt.title('Top 15 Neighborhoods – My Combined Score')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

### The Result

The winner.

```python
best = df_final.iloc[0]
print('Best Neighborhood (by my metric):', best['Neighborhood'])
print('Affordability Score:', round(best['Affordability Score'], 3))
print('Transit Score:      ', round(best['Transit Score'], 3))
print('Combined Score:     ', round(best['My Score'], 3))
```
